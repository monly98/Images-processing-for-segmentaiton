```python
#Average Filter
from matplotlib import pyplot as plt
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Median Filter
from google.colab.patches import cv2_imshow
median = cv2.medianBlur(img, 5)
#compare = np.concatenate((img, median), axis=1) #side by side comparison
compare = np.concatenate((img,median), axis=1) #side by side comparison
cv2_imshow( compare)
```

```python
#Min and Max Filter
from PIL import Image, ImageOps, ImageFilter
import cv2
# creating a image object
im1 = Image.open("brain.jpg")
im1 = ImageOps.grayscale(im1)
# applying the min filter
Minfilt = im1.filter(ImageFilter.MinFilter(size = 3))
Maxfilt = im1.filter(ImageFilter.MaxFilter(size = 3))   
plt.subplot(121),plt.imshow(Minfilt),plt.title('MinFilter Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Maxfilt),plt.title('MaxFilter Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
