```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# loading image
img0 = cv2.imread('gojo1.jpg')

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels, filters must be odd i.e ksize = 3, 5, 7, 9, 11 ...etc
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,dx=0, dy=1,ksize=3)  # x, horizontal
sobely = cv2.Sobel(img,cv2.CV_64F,dx=1, dy=0,ksize=3)  # y, vertical 
sobelxy = cv2.Sobel(img,cv2.CV_64F,dx=1, dy=1,ksize=3)  # xy, both

plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap='gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()

plt.subplot(2,2,1),plt.imshow(sobelxy,cmap = 'gray')  
plt.title('sobelxy'), plt.xticks([]), plt.yticks([])
# converting to its binary form
ret, bw_img = cv2.threshold(sobely, 127, 255, cv2.THRESH_BINARY)
#plt.imshow( bw_img,cmap='binary')
plt.subplot(2,2,2),plt.imshow(bw_img,cmap = 'binary')  
plt.title('sobelx binaryimage'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)
plt.subplot(121),plt.imshow(img_prewittx,cmap = 'gray')
plt.title('prewit x'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_prewitty,cmap = 'gray')
plt.title('prewit y'), plt.xticks([]), plt.yticks([])
plt.show()
plt.imshow(img_prewittx+img_prewitty,cmap='gray')
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('gojo1.jpg',0)
edges = cv.Canny(img,100,200)


plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
